```python
import pandas as pd
```

```python
vaccine = pd.read_csv("vaccination_data.csv")
vaccine.head()
```

```python
covid = pd.read_csv("covid_data.csv")
covid.head()
```

```python
vaccine.info()
```

```python
#Convert the Date Variable to Date Datatype

vaccine["DATE_UPDATED"] = pd.to_datetime(vaccine["DATE_UPDATED"])
vaccine.info()
```

```python
#Change column name if needed, "inplace = True" is important!

vaccine.rename(columns = {'PERSONS_VACCINATED_1PLUS_DOSE': 'DOSE_1',
                          'PERSONS_FULLY_VACCINATED': 'DOSE_2',
                          'PERSONS_BOOSTER_ADD_DOSE': 'BOOSTER'
                         }, inplace = True)
vaccine.info()
```

```python
#SELECT part of variable from the dataframe
#SELECT country, total_vaccination, dose 1, dose 2 and booster FROM vaccine
#Create new dataframe first, for example we used "df" the well-know variable used for "dataframe"

df = vaccine[['COUNTRY','TOTAL_VACCINATIONS','DOSE_1','DOSE_2','BOOSTER']]
df.head()
```

```python
#Create one column to add category of total vaccinations
import numpy as np

df['COUNT_CATEGORY'] = np.where(
    df['TOTAL_VACCINATIONS'] > 5000000, 'Huge', np.where(
    df['TOTAL_VACCINATIONS'] < 5000000, 'Small', -1))
df.head()
```

```python
#Create new DataFrame with dropped column "COUNT_CATEGORY"

df_dropped = df.drop('COUNT_CATEGORY', axis=1)
df_dropped.head()
```

```python
#Filtering Covid Data, select only Indonesia

covid[covid['Country'] == 'Indonesia']
```

```python
#Filtering Covid Data, select only Indonesia and Republic of Korea (South Korea)

covid[covid['Country'].isin(['Indonesia','Republic of Korea'])]
```

```python
#Filtering using part of the country keyword "Ko"

covid[covid['Country'].str.contains("Ko")]
```

```python
#Sometimes we need to pivot the data

#Convert the Date Variable to Date Datatypes
covid['Date_reported'] = pd.to_datetime(covid['Date_reported'])

#Transform date to Quarter
covid['Quarter'] = covid['Date_reported'].dt.to_period('Q')

#Pivoting
covid_pivot = covid.pivot_table(index='Country',
                               columns='Quarter',
                               values='New_cases',
                               aggfunc='sum')
covid_pivot.head(10)
```

```python
#GROUP BY function

covid_group1 = covid.groupby(by="Country").sum() #Just one group and SUM all data
covid_group1.head()
```

```python
#IF we need some adjustment, we can use AGGREGATE function

covid_group2 = covid.groupby('Country').agg({
    'New_cases': ['mean', 'max'],
    'New_deaths': 'sum'})
covid_group2.head()
```

```python
#We will use aggregate to gather sum of new cases and deaths

covid_group = covid.groupby('Country').agg({
    'New_cases': 'sum',
    'New_deaths': 'sum'}).reset_index()
covid_group.head()
```

```python
#Join Data

covid_vaccine = covid_group.merge(vaccine,
                                 left_on='Country',
                                 right_on='COUNTRY',
                                 suffixes=(False, False),
                                 how='inner')
covid_vaccine
```

```python
#a (What is the top 5 country that has the most death ratio? (find death-per-case)

#Group by death ratio country and use Aggregate function on Cumulative_deaths and Cumulative_case to find the sum of the cases
death_ratio = covid.groupby('Country').agg({
    'Cumulative_deaths':'max',
    'Cumulative_cases':'max'
}).reset_index()

#Create new column to store the Count death/case ratio and show the top 5 country that has the most death ratio
death_ratio['Death/case'] = death_ratio['Cumulative_deaths']/death_ratio['Cumulative_cases']
death_ratio.sort_values(by='Death/case', ascending = False).head()

#Answer: Yemen, Sudan, Syrian Arab Republic, Peru, Somalia
```

```python
#B
booster_ratio = vaccine[['COUNTRY', 'BOOSTER', 'TOTAL_VACCINATIONS']]

booster_ratio = booster_ratio.fillna(0)

df = vaccine.groupby('COUNTRY').agg({
    'BOOSTER':'max',
    'TOTAL_VACCINATIONS':'max'}).reset_index()

df['booster_ratio'] = df['BOOSTER']/df['TOTAL_VACCINATIONS']
df_sort = df.sort_values(by='booster_ratio', ascending = False)
df_sort.head()

#Answer : Iceland, Jersey, Gibraltar, Italy, Singapore
```

```python
mobility = pd.read_csv("mobility_indonesia.csv")
mobility.head()
```

```python
#Mengubah Data
mobility["date"] = pd.to_datetime(mobility["date"])
mobility.info()

#Filter Date dari awal April sampai akhir April
mobility = mobility[(mobility['date'] >= "2022-04-01") & (mobility['date'] <= "2022-04-30")]
```

```python
mobility.rename(columns = {'retail_and_recreation_percent_change_from_baseline': 'retail_and_recreation',
                          'grocery_and_pharmacy_percent_change_from_baseline': 'grocery_and_pharmacy',
                          'parks_percent_change_from_baseline': 'parks',
                          'transit_stations_percent_change_from_baseline': 'transit_stations',
                          'workplaces_percent_change_from_baseline': 'workplace',
                          'residential_percent_change_from_baseline': 'residential'
                         }, inplace = True)
mobility.info()
```

```python
#10 Provinsi dengan mean workplace terbesar pada bulan April 2022

mobility1 = mobility.groupby('sub_region_1').agg({
    'workplace':'mean'
}).reset_index()
mobility1.sort_values(by=['workplace'], ascending = False).head(10)

#Answer: East Nusa Tenggara, South East Sulawesi, Central Sulawesi, Jambi, Central Kalimantan, Lampung, West Sulawesi, Bengkulu, Central Java, Riau
```

```python
#10 Provinsi dengan median workplace terbesar pada bulan April 2022

mobility2 = mobility
mobility2
mobility2["baru"] = (mobility2['retail_and_recreation'] + 
                     mobility2['grocery_and_pharmacy'] +
                     mobility2['parks'] +
                     mobility2['transit_stations'] +
                     mobility2['workplace'] +
                     mobility2['residential'])/6

mobility2 = mobility2.groupby('sub_region_1').agg ({
    "baru": 'median',
    "grocery_and_pharmacy": 'median'
}).reset_index()
mobility2.sort_values(by=['baru'], ascending = False).head(10)
```

```python
#Provinsi Pulau Jawa dengan mobility grocery dan pharmacy (median) terbesar di bulan April 2022

pjawa = ['East Java', 'Central Java', 'West Java', 'Banten', 'Jakarta', 'Special Region of Yogyakarta']
mobility_pjawa = mobility2[mobility2["sub_region_1"].isin(pjawa)]

mobility_pjawa_grocery = mobility_pjawa.sort_values(by="grocery_and_pharmacy", ascending = False)
mobility_pjawa_grocery[["sub_region_1", "grocery_and_pharmacy"]].head()
```
